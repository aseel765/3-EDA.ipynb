```python
import pandas as pd

df = pd.read_csv("SuperMarket Analysis.csv")
df.head()
```

```python
df.duplicated().sum()
```

```python
df["Product line"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["Product line"].value_counts().plot(kind="bar")
plt.title("Sales Count by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### Insight 1:
Fashion accessories is the most sold product line.
Health and beauty has the lowest sales count.

```python
import matplotlib.pyplot as plt
```

```python
sales_by_product = df.groupby("Product line")["Sales"].sum()
sales_by_product.sort_values(ascending=False)
```

```python
sales_by_product.sort_values().plot(kind="barh")
plt.title("Total Sales by Product Line")
plt.xlabel("Total Sales")
plt.show()
```

### Insight 2:
Food and beverages generates the highest total sales revenue,
even though it is not the highest in transaction count.

```python
df["Gender"].value_counts()
```

```python
df["Gender"].value_counts().plot(kind="bar")
plt.title("Sales Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

### Insight 3:
Female customers have higher transaction counts than male customers.

```python
df.groupby("Gender")["Sales"].sum()
```

```python
df.groupby("Gender")["Sales"].sum().plot(kind="bar")
plt.title("Total Sales by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Sales")
plt.show()
```

### Insight 4:
Female customers not only have more transactions,
but also generate higher total sales revenue than male customers.

```python
import seaborn as sns

sns.scatterplot(x="Unit price", y="Rating", data=df)
plt.title("Unit Price vs Rating")
plt.show()
```

### Insight 5:
There is no clear relationship between unit price and customer rating.
Higher-priced products do not necessarily receive higher ratings.
